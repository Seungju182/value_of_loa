```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("휴학의 가치 측정.csv")
```

```python
data = data.drop(columns='Timestamp')
```

```python
data.columns = ['지침', '학기 마칠 자신', '휴식', '신체 불편', '자신감, 자존감', '재충전 필요도',
                '여행', '졸업 아까움', '하고 싶은 것', '색다른 경험', '경험 필요도',
                '시험', '인턴', '진로 불확실', '학회', '진로 필요도',
                '휴학 경험', '휴학 결정 이유', '휴학 안 한 이유'
               ]
```

```python
data.head()
```

```python
conv_dict={'전혀 필요하지 않다.': 0.,'필요하지 않다.': 1.,'보통이다.': 2.,'필요하다.':3.,'매우 필요하다.':4.}
```

```python
df = data.replace(conv_dict)
```

```python
df.head()
```

```python
charge_df = df.iloc[:, :6]
experience_df = df.iloc[:, 6:11]
career_df = df.iloc[:, 11:16]
```

# Linear Regression including Constant

```python
import statsmodels.api as sm

def regress(data):
    X_train = data[data.columns[:-1]]
    y = data[data.columns[-1]]
    X = sm.add_constant(X_train)
    est = sm.OLS(y, X).fit()
    return est.summary()
```

```python
regress(charge_df)
```

```python
regress(experience_df)
```

```python
regress(career_df)
```

# Linear Regression without constant term

```python
def regress_without_constant(data):
    X = data[data.columns[:-1]]
    y = data[data.columns[-1]]
    #X = sm.add_constant(X_train)
    est = sm.OLS(y, X).fit()
    return est.summary()
```

```python
regress_without_constant(charge_df)
```

```python
regress_without_constant(experience_df)
```

```python
regress_without_constant(career_df)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.tree import export_graphviz
from graphviz import Source


def decision_tree(data):
    model = DecisionTreeClassifier()
    X = data[data.columns[:-1]]
    y = data[data.columns[-1]]
    model.fit(X, y)
    y_pred = model.predict(X)
    acc = accuracy_score(y, y_pred)
    print("Accuracy score: ", acc)
    
    treedot = export_graphviz(model, feature_names = list(X.columns), class_names = list(map(str, sorted(y.unique()))),
                              max_depth = 4, impurity=False, rounded=True, special_characters=True, out_file=None)
    treegraph = Source(treedot, format="png")
    treegraph.render(f"{data.columns[-1]}_의사결정나무")
    return treegraph
```

```python
decision_tree(charge_df)
```

```python
decision_tree(experience_df)
```

```python
decision_tree(career_df)
```
